# Q1. Supervised Learning

**Task-1. Data Loading and Inspection**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Krishna-97/ml-assessment-krishna-tk/main/data/q1_heart_disease.csv')

```

*Displaying Shape, Data Types, Missing Values of df*

```python

print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

*Displaying the first 5 rows of df*

```python
print("\nFirst 5 rows:")
print(df.head())
```

**Task-2 Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#To set Aesthetic Style
sns.set(style="whitegrid")
```

*Visualization-1: Target class Distribution*

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='heart_disease', data=df,hue='heart_disease')
plt.title('Distribution of Heart Disease (Target Variable)')
plt.xlabel('Heart Disease (0 = Absent, 1 = Present)')
plt.ylabel('Count')
plt.savefig('Target_class_distribution.png')
print('Visualization Generated :Target_class_distribution.png')
plt.show()
```

*Visualization-2: Correlation HeatMap*

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig('Correlation_heatmap.png')
print('Visualization Generated :Correlation_heatmap.png')
plt.show()
```
